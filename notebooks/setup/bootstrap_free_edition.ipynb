```sql
%sql

-- Criar catálogo
CREATE CATALOG IF NOT EXISTS olist;

-- Atribuir catálogo
USE CATALOG olist;

-- Criar schemas (medalhão)
CREATE SCHEMA IF NOT EXISTS landing;
CREATE SCHEMA IF NOT EXISTS bronze;
CREATE SCHEMA IF NOT EXISTS silver;
CREATE SCHEMA IF NOT EXISTS gold;

-- Atribuir schema
USE SCHEMA landing;

-- Criar volumes
CREATE VOLUME IF NOT EXISTS files
COMMENT 'Arquivos brutos recebidos do dataset OLIST';

CREATE VOLUME IF NOT EXISTS checkpoints
COMMENT 'Auto Loader checkpoints';

CREATE VOLUME IF NOT EXISTS schema
COMMENT 'Auto Loader checkpoints';
```

```sql
%python

tables = [
    "customers",
    "order_items",
    "order_payments",
    "order_reviews",
    "orders",
    "products",
    "sellers"
]

base_path = "/Volumes/olist/landing/files"

for table in tables:
    dbutils.fs.mkdirs(f"{base_path}/{table}/")
```

```sql
%python
csv_source = '/Workspace/Users/gustavosass@gmail.com/CSVS'
data_root = '/Volumes/olist/landing/files'

csv_map = {
    'olist_customers_dataset.csv': 'customers',
    'olist_order_items_dataset.csv': 'order_items',
    'olist_order_payments_dataset.csv': 'order_payments',
    'olist_order_reviews_dataset.csv': 'order_reviews',
    'olist_orders_dataset.csv': 'orders',
    'olist_products_dataset.csv': 'products',
    'olist_sellers_dataset.csv': 'sellers'
}

for csv_file, folder in csv_map.items():
    src = f'{csv_source}/{csv_file}'
    dst = f'{data_root}/{folder}/{csv_file}'
    dbutils.fs.cp(src, dst)
    print(f'Copiado: {src} -> {dst}')
```
