```python
import shutil
from pathlib import Path

# Pasta de origem dos CSVs (absoluto, para garantir)
csv_source = Path('/home/jovyan/work/olist_files')

data_root = Path('./data/olist/landing/files')

csv_map = {
    'olist_customers_dataset.csv': 'customers',
    'olist_geolocation_dataset.csv': 'geolocation',
    'olist_order_items_dataset.csv': 'order_items',
    'olist_order_payments_dataset.csv': 'order_payments',
    'olist_order_reviews_dataset.csv': 'order_reviews',
    'olist_orders_dataset.csv': 'orders',
    'olist_products_dataset.csv': 'products',
    'olist_sellers_dataset.csv': 'sellers'
}

for csv_file, folder in csv_map.items():
    src = csv_source / csv_file
    dst_folder = data_root / folder
    dst_folder.mkdir(parents=True, exist_ok=True)
    dst = dst_folder / csv_file
    if src.exists():
        shutil.copy(src, dst)
        print(f'Copiado: {src} -> {dst}')
    else:
        print(f'Arquivo não encontrado: {src}')
```
